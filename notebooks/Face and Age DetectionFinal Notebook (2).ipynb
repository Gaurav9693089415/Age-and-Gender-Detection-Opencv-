##  Age and Gender Detection using CNN

This project uses the UTKFace dataset to build a Convolutional Neural Network (CNN) model that performs **multi-output learning** — predicting both a person's **age** (regression) and **gender** (classification) from a facial image.


## Importing all necessary libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from PIL import Image

from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

## Data Loading and Preprocessing

Each image filename contains encoded age and gender information in the format: `age_gender_race_date.jpg`. We extract these labels and preprocess the images (resizing, normalization) before feeding them to the model.


```python
# Paths
image_dir = r'C:\Users\patha\FaceAgingApp\data\Real&UTKFace\\'  #path where we have stored all our image dataset


# =========================
# Load and Preprocess Images
# =========================
def extract_label(filename):
    parts = filename.split('_')
    age = int(parts[0])       # Age is first part
    gender = int(parts[1])    # Gender is second part
    return age, gender

# Initialize lists
images = []
ages = []
genders = []
```

## To Show Sample Images

Let's display a few sample images to ensure they are loaded and labeled corre
ctly.


```python
# Display 5 sample images with age and gender
for i in range(5):
    plt.imshow(images[i])
    label = "Male" if genders[i] == 0 else "Female"
    plt.title(f"Age: {ages[i]}, Gender: {label}")
    plt.axis('off')
    plt.show()
```

## Preparing the Data


We one-hot encode the gender labels and split the dataset into training and testing sets.


```python
# Process all images
image_files = os.listdir(image_dir)

for image_file in tqdm(image_files, desc="Loading images"):
    if image_file.endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(image_dir, image_file)
        
        # Load and resize
        img = load_img(img_path, target_size=(224, 224))
        img_array = img_to_array(img) / 255.0  # Normalize
        images.append(img_array)
        
        # Extract labels
        age, gender = extract_label(image_file)
        ages.append(age)
        genders.append(gender)

# Convert to numpy arrays
images = np.array(images)
ages = np.array(ages)
genders = np.array(genders)

# Gender: One-hot encode
genders = to_categorical(genders, num_classes=2)

# Train-test split
X_train, X_test, y_train_age, y_test_age, y_train_gender, y_test_gender = train_test_split(
    images, ages, genders, test_size=0.2, random_state=42
)
```

## Building  CNN Model

The model uses convolutional layers and shares features for both outputs:
- **Age** prediction (regression)
- **Gender** prediction (classification)


```python
# Build Model
# 
input_layer = Input(shape=(224, 224, 3))
x = Conv2D(32, (3, 3), activation='relu')(input_layer)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu')(x)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)

# Two Outputs
age_output = Dense(1, name='age_output')(x)                        # Regression (age)
gender_output = Dense(2, activation='softmax', name='gender_output')(x)  # Classification (gender)

model = Model(inputs=input_layer, outputs=[age_output, gender_output])
```

## Compiling the Model

We use different loss functions for each output:
- Mean Squared Error (MSE) for age
- Categorical Crossentropy for gender


```python
# Compile Model
model.compile(
    optimizer=Adam(),
    loss={'age_output': 'mse', 'gender_output': 'categorical_crossentropy'},
    metrics={'age_output': 'mae', 'gender_output': 'accuracy'}
)

model.summary()
```

## Model Training

We use early stopping with 3 patience to prevent overfitting and train the model for up to 20 epochs.


```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Early stopping callback to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)


# =========================
history = model.fit(
    X_train,
    {'age_output': y_train_age, 'gender_output': y_train_gender},
    validation_data=(X_test, {'age_output': y_test_age, 'gender_output': y_test_gender}),
    epochs=20,
    batch_size=32,
    callbacks=[early_stopping]  
)
```

## Evaluating  the model
Evaluating  the model on the testing dataset


```python
# Evaluate the model on the test set
evaluation = model.evaluate(
    X_test,
    {'age_output': y_test_age, 'gender_output': y_test_gender},
    batch_size=32
)

# Print evaluation results
print("\nEvaluation Results:")
print(f"Total Loss: {evaluation[0]}")
print(f"Age Loss: {evaluation[1]}")
print(f"Gender Loss: {evaluation[2]}")
print(f"Age MAE (Mean Absolute Error): {evaluation[3]}")  # assuming you used 'mae' metric for age
print(f"Gender Accuracy: {evaluation[4]}")  # for gender classification
```

## Saving the model
 Saving the model as age_gender_model.keras

```python
# Save Model
model.save('age_gender_model2.keras')
print("Model saved as 'age_gender_model2.keras'")
```

## Predict Age and Gender from an Input Image

The user provides a path to an image. The image is then preprocessed, passed through the model, and the predictions are displayed.


```python
# Import Dependencies 
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array

#Load the Pre-trained Model
model = load_model('age_gender_model2.keras')  
print(" Model loaded successfully.")

# ====== Step 1: Ask for Image Path ======
user_image_path = input("Enter the full path to your image: ").strip('"').strip("'")

# Check if the file exists
if not os.path.exists(user_image_path):
    print(f" Error: The image file at {user_image_path} does not exist.")
    exit()

# ====== Step 2: Preprocess the Image ======
def preprocess_image(image_path, target_size=(224, 224)):
    img = load_img(image_path, target_size=target_size)
    img_array = img_to_array(img)
    img_array = img_array / 255.0  # Normalize pixel values
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array

processed_image = preprocess_image(user_image_path)

# Step 3: Make Predictions
predictions = model.predict(processed_image)

# Age prediction
predicted_age = float(predictions[0][0])  # Assuming first output is age

# Gender prediction
predicted_gender_probs = predictions[1][0]  # Assuming second output is gender probabilities
predicted_gender = np.argmax(predicted_gender_probs)  # 0 for Male, 1 for Female
gender_str = "Male" if predicted_gender == 0 else "Female"

# Step 4: Display Results
print("\n===== Prediction Results =====")
print(f" Predicted Age: {predicted_age:.1f} years")
print(f"Predicted Gender: {gender_str}")

# Step 5: Show Image with Title
img = load_img(user_image_path)
plt.imshow(img)
plt.title(f"Age: {predicted_age:.1f} yrs, Gender: {gender_str}")
plt.axis('off')
plt.show()
```
